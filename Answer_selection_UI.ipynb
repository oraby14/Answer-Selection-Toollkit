```python
from keras.models import load_model
import pandas as pd
import numpy as np
from Data_prep import question_preprocessing
from tkinter import *
class MyWindow:
    def __init__(self, win):
        self.lbl1=Label(win, text='The question')
        self.lbl2=Label(win, text='Asnwer 1')
        self.lbl3=Label(win, text='Asnwer 2')
        self.lbl4=Label(win, text='Asnwer 3')
        self.lbl5=Label(win, text='Asnwer 4')
        self.lbl6=Label(win, text='The Result')
        self.t1=Entry()
        self.t2=Entry()
        self.t3=Entry()
        self.t4=Entry()
        self.t5=Entry()
        self.t6=Entry(width = 45)
        self.lbl1.place(x=100, y=50)
        self.t1.place(x=200, y=50)
        self.lbl2.place(x=100, y=100)
        self.t2.place(x=200, y=100)
        self.lbl3.place(x=100, y=150)
        self.t3.place(x=200, y=150)
        self.lbl4.place(x=100, y=200)
        self.t4.place(x=200, y=200)
        self.lbl5.place(x=100, y=250)
        self.t5.place(x=200, y=250)
        self.btn1 = Button(win, text='Predict')
        self.b1=Button(win, text='Predict', command=self.Predict)
        self.btn2 = Button(win, text='Match')
        self.b2=Button(win, text='Match', command=self.solve)
        self.b1.place(x=150, y=300)
        self.b2.place(x=250, y=300)
        self.lbl6.place(x=20, y=350)
        self.t6.place(x=100, y=350)
   
    def Predict(self):
        self.t6.delete(0, 'end')
        question = self.t1.get()
        answer_1 = self.t2.get()
        answer_2 = self.t3.get()
        answer_3 = self.t4.get()
        answer_4 = self.t5.get()
        model = load_model('answer_selection.h5')
        a = pd.DataFrame()
        a[0] = [question]
        a = pd.DataFrame(np.repeat(a.values, 4, axis=0))
        a.insert(1, "1", [answer_1, answer_2, answer_3,answer_4 ], True)
        pre_ans , pre_ques =question_preprocessing(a)
        y_pred = model.predict([pre_ans,pre_ques])
        self.t6.insert(END, (y_pred))
        highst_rank = y_pred[0]
        for number in highst_rank:
            if number > highst_rank:
                highst_rank = number
        index = np.where(y_pred == highst_rank )
    def solve(self):
        self.t6.delete(0, 'end')
        split_question = set(self.t1.get().split())
        split_answer_1 = set(self.t2.get().split())
        split_answer_2 = set(self.t3.get().split())
        split_answer_3 = set(self.t4.get().split())
        split_answer_4 = set(self.t5.get().split())
        
        common_words_1 = split_question & split_answer_1
        common_words_2 = split_question & split_answer_2
        common_words_3 = split_question & split_answer_3
        common_words_4 = split_question & split_answer_4
        
        if (len(common_words_1) >= len(common_words_2)) and (len(common_words_1) >= len(common_words_3)) and (len(common_words_1) >= len(common_words_4)) :
            self.t6.insert(END, ("answer 1 is the right"))
        elif (len(common_words_2) >= len(common_words_1)) and (len(common_words_2) >= len(common_words_3)) and (len(common_words_2) >= len(common_words_4)) :
            self.t6.insert(END, ("answer 2 is the right"))
        elif (len(common_words_3) >= len(common_words_1)) and (len(common_words_3) >= len(common_words_2)) and (len(common_words_3) >= len(common_words_4)) :
            self.t6.insert(END, ("answer 3 is the right"))  
        else:
            largest = len(common_words_3)
            self.t6.insert(END, ("answer 4 is the right"))
        
        
        
        
        
window=Tk()
mywin=MyWindow(window)
window.title('ANSWER SELECTION TOOLKIT')
window.geometry("400x400+10+10")
window.mainloop()
```
